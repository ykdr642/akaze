```python
def getScore(a,prime,target):
    sxd = int(a[prime][0][0]) - int(a[target][0][0])
    syd = int(a[prime][0][1]) - int(a[target][0][1])
    bxd = int(a[prime][1][0]) - int(a[target][1][0])
    byd = int(a[prime][1][1]) - int(a[target][1][1])
    if prime is not target:
        if bxd != 0:
            xration = sxd / bxd
        else:
            xration = None
        if byd != 0:
            yration = syd / byd
        else:
            yration = None
        #print(str(target)+":",xration,yration)
    else:
        xration = None
        yration = None
    return [xration,yration]
```

```python
def getave(grid,xN,yN):
    xscore = []
    yscore = []
    if len(grid[xN][yN]) > 1:
        for i in range(len(grid[xN][yN])):
            x,y = getScore(grid[xN][yN],0,i)
            xscore.append(x)
            yscore.append(y)
        xave = 0
        yave = 0
        xcount = 0
        ycount = 0
        for x in xscore:
            if x is not None:
                xave += x
                xcount += 1
        for y in yscore:
            if y is not None:
                yave += y
                ycount += 1
        #print("xAverage:",xave/xcount)
        #print("yAverage:",yave/ycount)
        return [xave/xcount,yave/ycount]
    else:
        return [None,None]
```

```python
img2 = cv2.imread('yoko.PNG')
img1 = cv2.imread('tate.PNG')

y1 = 600
y2 = 700
x1 = 420
x2 = 670
Rimg1 = img1[y1:y2,x1:x2]

# A-KAZE検出器の生成
detector = cv2.AKAZE_create()

# 特徴量の検出と特徴量ベクトルの計算
kp1, des1 = detector.detectAndCompute(Rimg1, None)
kp2, des2 = detector.detectAndCompute(img2, None)

# Brute-Force Matcherの生成
bf = cv2.BFMatcher()

# 特徴量ベクトル同士をBrute-Force＆KNNでマッチング
matches = bf.knnMatch(des1, des2, k=2)

# データを間引く
ratio = 0.2
good = []
for m, n in matches:
    if m.distance < ratio * n.distance:
        good.append([m])

# 特徴量をマッチング状況に応じてソート
good = sorted(matches, key = lambda x : x[1].distance)

# 対応する特徴点同士を描画
img3 = cv2.drawMatchesKnn(Rimg1, kp1, img2, kp2, good[:2], None, flags=2)

cv2.imshow('img', img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def getPoint(X,Y,matchPoint):
    matchPoint.sort(key=lambda x:(x[0][0]-X)**2+(x[0][1]-Y)**2)
    for i in range(len(matchPoint)):
        if matchPoint[0][0][0] != matchPoint[i][0][0] and matchPoint[0][1][0] != matchPoint[i][1][0] and \
            matchPoint[0][0][1] != matchPoint[i][0][1] and matchPoint[0][1][1] != matchPoint[i][1][1]:
            xp = getTarget(X,matchPoint[0][0][0],matchPoint[i][0][0],matchPoint[0][1][0],matchPoint[i][1][0])
            yp = getTarget(Y,matchPoint[0][0][1],matchPoint[i][0][1],matchPoint[0][1][1],matchPoint[i][1][1])
            break
    return (int(xp),int(yp))
```
